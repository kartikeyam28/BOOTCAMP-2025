```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
df = pd.read_csv('bright_automotive_company.csv')

# Boxplot to check for outliers
df.plot(kind='box', subplots=True, layout=(2, 3), figsize=(12, 8), sharex=False, sharey=False)
plt.suptitle("Boxplots(Outlier Detection)")
plt.show()
numeric_df = df.select_dtypes(include=['number'])
iqr = numeric_df.quantile(0.75) - numeric_df.quantile(0.25)
lower_bound = numeric_df.quantile(0.25) - (1.5 * iqr)
upper_bound = numeric_df.quantile(0.75) + (1.5 * iqr)
outliers = ((numeric_df < lower_bound) | (numeric_df > upper_bound)).any(axis=1)
print("Number of outliers:", outliers.sum())
print("Percentage of outliers:", (outliers.sum() / len(df)) * 100)
print("Outliers:")
print(df[outliers])
print("no. of outliers:", numeric_df>upper_bound)
print("no. of outliers:", numeric_df<lower_bound)
outlier_counts = {}
for col in numeric_df.columns:
    col_outliers = ((numeric_df[col] < lower_bound[col]) | (numeric_df[col] > upper_bound[col]))
    outlier_counts[col] = col_outliers.sum()
    if col_outliers.sum() > 0:
        print(f"Outliers in column '{col}':")
        print(df.loc[col_outliers, col])

print("\nIndividual number of outliers per column:")
for col, count in outlier_counts.items():
    print(f"'{col}': {count}")



# Countplots for categorical features (if any)
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    plt.figure(figsize=(6, 3))
    sns.countplot(x=col, data=df)
    plt.title(f"Countplot for {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Heatmap for correlation
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Is there a correlation between age and salary? Provide the correlation coefficient and interpret the result.
```

```python
import matplotlib.pyplot as plt
# Calculate the correlation coefficient between 'Age' and 'Salary'
correlation_coefficient = df['Age'].corr(df['Salary'])

print(f"The correlation coefficient between Age and Salary is: {correlation_coefficient:.2f}")

# Interpret the result
if correlation_coefficient > 0.5:
    interpretation = "There is a strong positive correlation between Age and Salary. As age increases, salary tends to increase significantly."
elif correlation_coefficient > 0.2:
    interpretation = "There is a moderate positive correlation between Age and Salary. As age increases, salary tends to increase."
elif correlation_coefficient > 0:
    interpretation = "There is a weak positive correlation between Age and Salary. There is a slight tendency for salary to increase with age."
elif correlation_coefficient < -0.5:
    interpretation = "There is a strong negative correlation between Age and Salary. As age increases, salary tends to decrease significantly."
elif correlation_coefficient < -0.2:
    interpretation = "There is a moderate negative correlation between Age and Salary. As age increases, salary tends to decrease."
elif correlation_coefficient < 0:
    interpretation = "There is a weak negative correlation between Age and Salary. There is a slight tendency for salary to decrease with age."
else:
    interpretation = "There is little to no linear correlation between Age and Salary."

print(f"Interpretation: {interpretation}")

# Optional: Visualize the correlation with a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title("Scatter Plot of Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("bright_automotive_company.csv")
df['Partner_working'] = df['Partner_working'].str.lower().str.strip()
df['House_loan'] = df['House_loan'].str.lower().str.strip()
df['Personal_loan'] = df['Personal_loan'].str.lower().str.strip()
df['Make'] = df['Make'].str.upper().str.strip()
df['Profession'] = df['Profession'].str.title().str.strip()

salary_comparison = df.groupby('Partner_working')['Salary'].mean()
house_loan_proportion = df[df['House_loan'] == 'yes'].groupby('Profession').size() / df.groupby('Profession').size()

sns.boxplot(data=df, x='Personal_loan', y='Salary')
plt.title("Salary Distribution by Personal Loan Status")
plt.tight_layout()
plt.show()

automobile_salary = df.groupby('Make')['Salary'].mean()
avg_price = df['Price'].mean()
price_by_salary = df.groupby(pd.cut(df['Total_salary'], bins=4))['Price'].mean()

print("Salary Comparison:\n", salary_comparison)
print("\nHouse Loan Proportion by Profession:\n", house_loan_proportion)
print("\nAverage Salary by Automobile Make:\n", automobile_salary)
print("\nAverage Price of Product/Service:", avg_price)
print("\nPrice Variation by Total Salary Ranges:\n", price_by_salary)
```

```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv("bright_automotive_company.csv")

# Ensure House_loan column is processed to lowercase before mapping
df['House_loan'] = df['House_loan'].str.lower().str.strip()

# Map lowercase 'yes' and 'no' to 1 and 0
df['House_loan_binary'] = df['House_loan'].map({'yes': 1, 'no': 0})

# Ensure Education column is processed as well for consistent grouping
df['Education'] = df['Education'].str.title().str.strip()


house_loan_by_education = df.groupby('Education')['House_loan_binary'].mean()
print("House Loan Rates by Education Level:")
print(house_loan_by_education)

# Drop rows where House_loan_binary is NaN before creating the contingency table
# This handles cases where the original 'House_loan' had unexpected values
contingency_edu_loan = pd.crosstab(df['Education'], df['House_loan_binary']).dropna(axis=1, how='all').dropna(axis=0, how='all')


# Check if the contingency table is empty before performing the chi-squared test
if contingency_edu_loan.empty:
    print("\nError: Contingency table is empty. Check data or mapping.")
else:
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_edu_loan)

    print("\nChi-Squared Test:")
    print(f"Chi2 Statistic = {chi2}")
    print(f"P-value = {p_value}")
```

```python
import pandas as pd
import scipy.stats as stats

df['Personal_loan_binary'] = df['Personal_loan'].map({'Yes': 1, 'No': 0})
loan_grouped_salary = df.groupby('Personal_loan_binary')['Total_salary'].mean()
t_stat, p_val = stats.ttest_ind(
    df[df['Personal_loan_binary'] == 1]['Total_salary'].dropna(),
    df[df['Personal_loan_binary'] == 0]['Total_salary'].dropna(),
    equal_var=False
)
print(loan_grouped_salary)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```
